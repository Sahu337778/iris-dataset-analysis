# Model Building and Evaluation:

```python
#Divide the feature and target variable

x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3, random_state=1)
```

# Scaling and Model Optimization:

Features are scaled using StandardScaler to improve model performance.

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```



The results obtained were not satisfactory as our model wasn't optimized by scaling the data. Therefore, we need to perform hyperparameter tuning to enhance the model's performance. Let's proceed with the tuning process.

```python
parameters = {
    "criterion": ["gini", "entropy"],
    "max_depth": list(range(1,20)),
    "min_samples_leaf": list(range(1,20))
}
```

```python
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(DecisionTreeClassifier(), parameters, verbose=2)
grid.fit(xtrain, ytrain)
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
dt = mymodel(grid.best_estimator_)
```

GridSearchCV is employed to hyper-tune the Decision Tree Classifier, resulting in an optimized model with lower bias and variance.

Currently, we are constructing an optimized Decision Tree model with low bias and variance for improved performance.
